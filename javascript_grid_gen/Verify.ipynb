```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import json
```

```python
# CUDA version
g0 = json.load(open('./11gs/g0_fixed.json'))
g0 = np.array(g0).reshape((1, 9, 24, 24, 24))

# mol_gridify2
mg2 = json.load(open('./11gs/mol_gridify2.json'))
mg2 = np.array(mg2).reshape((1, 9, 24, 24, 24))
```

```python
np.all(np.isclose(g0, mg2))
```

```python
def compare_grids(g0, g1):
    plt.figure(figsize=(30,30))

    for i in range(24):
        for j in range(9):
            a = g0[0][j][i]
            b = g1[0][j][i]

            plt.subplot(24,18,(i*18)+(j*2)+1)
            plt.imshow(a)
            plt.subplot(24,18,(i*18)+(j*2)+2)
            plt.imshow(b)
```

```python
compare_grids(g0, mg2)
```
